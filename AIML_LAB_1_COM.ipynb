```python
import matplotlib.pyplot as plt
people = ['teja','sowkya','priya','vyshu','neha','mounika','navya','harini','asmaja','abhireddy','vishnu','bharath']
age = [18,19,17,16,19,20,18,19,20,18,19,16]
weight = [35,39,40,42,44,43,46,37,46,44,50,55]
height = [160,155,156,171,165,189,177,197,186,167,155,176]
plt.scatter(weight,height)
plt.title("relationship between weight and height of the students")
plt.ylabel('weight')
plt.xlabel('height')
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("people's weight in kgs",fontsize=12,fontstyle='italic')
plt.bar(x=people,height=weight,width=0.5,color='pink',edgecolor='k',alpha=0.6)
plt.xlabel("people",fontsize=12)
plt.ylabel("weight (in kgs)",fontsize=12)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='pink',edgecolor='k',alpha=0.6,bins=6)
plt.title("histogram of students weights",fontsize=16)
plt.xlabel("weights in kgs",fontsize=12)
plt.show()
```
